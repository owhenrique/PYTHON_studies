```python
import pandas as pd
```

```python
dataframe = pd.read_csv(r'data/telecom_users.csv')
display(dataframe)
```

```python
print(dataframe.info())
```

```python
dataframe = dataframe.drop('Unnamed: 0', axis=1)

dataframe['TotalGasto'] = pd.to_numeric(dataframe['TotalGasto'], errors='coerce')

dataframe = dataframe.dropna(how='all', axis=1)

dataframe = dataframe.dropna(how='any', axis=0)

display(dataframe)
```

```python
dataframe.loc[dataframe['Genero'] == 'Masculino', 'Genero'] = int(0)
dataframe.loc[dataframe['Genero'] == 'Feminino', 'Genero'] = int(1)
display(dataframe)
```

```python
for column in dataframe:
    dataframe.loc[dataframe[column] == 'Sim', column] = int(1)
    dataframe.loc[dataframe[column] == 'Nao', column] = int(0)
    dataframe.loc[dataframe[column] == 'SemInternet', column] = int(3)
    
dataframe.loc[dataframe['TipoContrato'] == '2 anos', 'TipoContrato'] = int(24)
dataframe.loc[dataframe['TipoContrato'] == 'Mensal', 'TipoContrato'] = int(1)
dataframe.loc[dataframe['TipoContrato'] == 'Anual', 'TipoContrato'] = int(12)

dataframe.loc[dataframe['FormaPagamento'] == 'CartaoCredito', 'FormaPagamento'] = int(0)
dataframe.loc[dataframe['FormaPagamento'] == 'DebitoAutomatico', 'FormaPagamento'] = int(1)
dataframe.loc[dataframe['FormaPagamento'] == 'BoletoEletronico', 'FormaPagamento'] = int(3)

dataframe.loc[dataframe['ServicoInternet'] == 'Fibra', 'ServicoInternet'] = int(5)
dataframe.loc[dataframe['ServicoInternet'] == 'DSL', 'ServicoInternet'] = int(6)

for column in dataframe:
    dataframe[column] = pd.to_numeric(dataframe[column], errors='ignore')
 
display(dataframe)
```

```python
print(dataframe.info())
```

```python
from sklearn.model_selection import train_test_split

x = dataframe[['Genero', 'Aposentado', 'Casado', 'Dependentes', 'MesesComoCliente', 'ServicoTelefone', '']]
y = dataframe['Churn']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

# Escolhendo dois métodos para ensinar a IA (regressão linear e árvore de decisão)
model_LinearRegression = LinearRegression()
model_DecisionThree = RandomForestRegressor()

# treinando a IA
model_LinearRegression.fit(x_train, y_train)
model_DecisionThree.fit(x_train, y_train)

# fazer previsão nos testes 

prevision_LR = model_LinearRegression.predict(x_test)
prevision_DT = model_DecisionThree.predict(x_test)

# r² (r2_score) diz quanto a previsão chegou perto do resultado real 

from sklearn.metrics import r2_score

print(r2_score(y_test, prevision_LR))
print(r2_score(y_test, prevision_DT))
```
